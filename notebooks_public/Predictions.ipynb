## Clean other stocks

```python
from dasai.stocks.load_stocks import *
import pandas as pd

for symbol in other_symbols:
    input_file = '..' / get_raw_data_path() / f'{symbol}.csv'
    output_file = '..' / get_cleaned_data_path() / f'{symbol}.parquet'
    convert_to_cleaned_data(input_file, output_file)
```

## Predict and save results

```python
from dasai.stocks.prediction import *

symbols = ['AAPL', 'NFLX', 'KO', 'IBM', 'MSFT']
for symbol in symbols:
    df = pd.read_parquet('..' / get_cleaned_data_path() / f'{symbol}.parquet')
    output_file = '..' / get_result_data_path() / f'{symbol}_prediction.csv'
    predict_stock_price(df, output_file)
```

```python
import matplotlib.pyplot as plt

def plot_prediction(all_df, pred_df, model):
    fig = model.plot(pred_df)
    fig.gca().plot(all_df.reset_index()['date'], all_df['adjusted_close'], 'k.')
```

## Apple

```python

apple_all_df = pd.read_parquet('..' / get_cleaned_data_path() / 'AAPL.parquet')

df = pd.read_parquet('..' / get_cleaned_data_path() / 'AAPL.parquet')
output_file = '..' / get_result_data_path() / 'AAPL_prediction.csv'
apple_model, apple_df = predict_stock_price(df, output_file, train_quantile=0.95)

plot_prediction(apple_all_df, apple_df, apple_model)
```
